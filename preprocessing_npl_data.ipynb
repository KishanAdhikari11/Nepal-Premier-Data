```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
npl_data = pd.read_csv('Nepal_Premier_League(1).csv')
npl_data.head()
```

```python
npl_data.isnull().sum()
```

```python
npl_data['Toss Winner'] = npl_data['Toss'].str.split(',', expand=True)[0]
npl_data['Toss Decision'] = npl_data['Toss'].str.extract(r'elected to (\w+ first)')[0]
```

```python
npl_data['Toss Winner'] = npl_data['Toss Winner'].str.replace(r'\s*\(NPL\)', '', regex=True)
```

```python
npl_data['Toss Winner']
```

```python
npl_data['Toss Decision']
```

```python
npl_data['Umpires'] = npl_data['Umpires'].str.strip("[]").str.replace("'", "").str.split(', ')
npl_data[['Umpire 1', 'Umpire 2']] = pd.DataFrame(npl_data['Umpires'].tolist(), index=npl_data.index)
npl_data.drop(columns=['Umpires'], inplace=True)
```

```python
npl_data[['Umpire 1', 'Umpire 2']]
```

```python
npl_data.drop(['Match Info','Points', 'Reserve Umpire', 'Series', 'Match days', 'Toss'],axis=1,inplace=True)
```

```python
npl_data.sample()
```

```python
npl_data['Team 1'] = npl_data['Team 1'].str.strip("(NPL)").str.replace("'", "").str.split(', ')
```

```python
npl_data['Team 2'] = npl_data['Team 2'].str.strip("(NPL)").str.replace("'", "").str.split(', ')
```

```python
npl_data['Team 1'] = npl_data['Team 1'].fillna('').astype(str)

# Apply the string operations to clean the column
npl_data['Team 1'] = (npl_data['Team 1']
                       .str.strip("[]")  # Remove square brackets
                       .str.replace("'", "")  # Remove single quotes
                       .str.replace("(NPL)", "")  # Remove the "(NPL)" substring
                       .str.strip())  # Remove leading/trailing spaces
```

```python
npl_data['Team 2'] = npl_data['Team 2'].fillna('').astype(str)

# Apply the string operations to clean the column
npl_data['Team 2'] = (npl_data['Team 2']
                       .str.strip("[]")  # Remove square brackets
                       .str.replace("'", "")  # Remove single quotes
                       .str.replace("(NPL)", "")  # Remove the "(NPL)" substring
                       .str.strip())  # Remove leading/trailing spaces
```

```python
npl_data['won_by_margin'] = npl_data['won_by_margin'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
```

```python
npl_data.head(3)
```

```python
npl_data.columns = npl_data.columns.str.replace(' ', '_').str.lower()
```

```python
npl_data.isnull().sum()
```

```python
npl_data = npl_data.replace('umbini Lions', 'Lumbini Lions')
```

```python
npl_data = npl_data.replace('okhara Avengers', 'Pokhara Avengers')
```

```python
npl_data = npl_data.replace('S', 'Sudur Paschim Royals')
```

```python
npl_data[['team_1', 'team_2', 'season', 'player_of_the_match', 'toss_winner', 'toss_decision', 'target', 'winner_team', 'won_by_margin' ,'umpire_1', 'umpire_2', 'tv_umpire', 'match_referee']]
```

```python
# Replace NaN values in the 'winner_team' column with 'Pokhra Avengers'
npl_data['winner_team'] = npl_data['winner_team'].fillna('Pokhra Avengers')
```

```python
npl_data['won_by_margin'] = npl_data['won_by_margin'].fillna('0 super over')
```

```python
updated_npl_data = npl_data[['team_1', 'team_2', 'season', 'player_of_the_match', 'toss_winner', 'toss_decision', 'target', 'winner_team', 'won_by_margin' ,'umpire_1', 'umpire_2', 'tv_umpire', 'match_referee']]
```

```python
updated_npl_data.to_csv('Nepal_Premier_League.csv')
```
